```python
import requests
from bs4 import BeautifulSoup
```

```python
def getdata(site_url):
    res=requests.get(site_url)
    return res.text
```

```python
def gethtml(site_url):
    data=getdata(site_url)
    soup=BeautifulSoup(data,'html.parser')
    return soup
```

```python
def getcustname(soup):
    data_string=""
    cust_list=[]
    for item in soup.find_all("p",class_="_2sc7ZR _2V5EHH"):
        data_string=data_string+item.get_text()
        cust_list.append(data_string)
        data_string=""
    return cust_list
```

```python
cust_names=getcustname(soup)
cust_names
```

```python
def getreviews(soup):
    data_string=""
    review_list=[]
    for item in soup.find_all("div",class_="t-ZTKy"):
        data_string=data_string+item.get_text()
        review_list.append(data_string)
        data_string=""
    return review_list
```

```python
reviews=getreviews(soup)
reviews
```

```python
def getrate(soup):
    data_string=""
    ratings_list=[]
    for item in soup.find_all("div",class_="_3LWZlK _1BLPMq"):
        data_string=data_string+item.get_text()
        ratings_list.append(data_string)
        data_string=""
    return ratings_list
```

```python
ratings=getrate(soup)
ratings
```
